```python
import pymongo
import charts
import re
```

```python
client = pymongo.MongoClient('localhost', 27017)
ganji = client['sample']
item_info = ganji['sample']
```

```python
for item in item_info.find():
    if re.search(r'\d+',item['price']):
        price = float(re.search(r'\d+',item['price']).group())
    else:
        price = 0
    item_info.update_one({'_id':item['_id']},{'$set':{'price':price}})
```

```python
def gen_data(area):
    pipeline = [
        {'$match':{'area':area}},
        {'$group':{'_id':'$cates','counts':{'$sum':1}}}, #cates已经被整理成一个内容了，不用$slice了
        {'$sort':{'counts':-1}},
        {'$limit':3}
    ]
    for i in item_info.aggregate(pipeline):
        data = {
            'name': i['_id'],
            'data': [i['counts']],
            'type': 'column'
        }
        yield data
```

```python
options = {
    'chart': {'type': 'zoomXY'},
    'title': {'text': '朝阳区发帖量Top3类目'},
    'subtitle': {'text': '发帖量可视化'}
}
series = [i for i in gen_data('朝阳')]
charts.plot(series,options=options,show='inline')
```

```python
def gen_data(cate):
    pipeline = [
        {'$match':{'$and':[{'cates':cate},
                          {'look':{'$nin':['-','${info.paramsMap.oldlevel}']}}]}},
        {'$group':{'_id':'$look','avgs':{'$avg':'$price'}}}, 
        {'$sort':{'avgs':-1}}
    ]
    for i in item_info.aggregate(pipeline):
        cates = i['_id']
        data = i['avgs']
        yield cates, data
```

```python
for i in gen_data('二手手机'):
    print(i)
```

```python
def gen_line(cate):
    pipeline = [
        {'$match':{'$and':[{'cates':cate},
                          {'look':{'$nin':['-','${info.paramsMap.oldlevel}']}}]}},
        {'$group':{'_id':'$look','avgs':{'$avg':'$price'}}}, 
        {'$sort':{'avgs':-1}}
    ]
    for i in item_info.aggregate(pipeline):
        data = {
            'name':cate,
            'data':[i[-1] for i in gen_data(cate)],
            'type':'line'
        }
        yield data
```

```python
options = {
    'title': {'text': '二手手机价格趋势'},
    'xAxis': {'categories': [i[0] for i in gen_data('二手手机')]},
    'yAxis': {'title': {'text': '平均价格'}}
}
series = [i for i in gen_line('二手手机')]
charts.plot(series,options=options,show='inline')
```

```python
options = {
    'title': {'text': '二手家电价格趋势'},
    'xAxis': {'categories': [i[0] for i in gen_data('二手家电')]},
    'yAxis': {'title': {'text': '平均价格'}}
}
series = [i for i in gen_line('二手家电')]
charts.plot(series,options=options,show='inline')
```
