```python
import pymongo
from string import punctuation
import charts
```

```python
client = pymongo.MongoClient('localhost',27017)
ganji = client['sample']
item_info = ganji['sample']
```

```python
for item in item_info.find().limit(300):
    print(item['cates'])
```

```python
for i in item_info.find():
    cate = i['cates'][2].split('北京')[-1]
#     print(cate)
    item_info.update({'_id':i['_id']},{'$set':{'cates':cate}})
```

```python
for item in item_info.find().limit(300):
    print(item['cates'])
```

```python
for item in item_info.find().limit(300):
    print(item['area'])
```

```python
for i in item_info.find():
    if i['area']:
        area = [i for i in i['area'] if i not in punctuation]
    else:
        area = ['不明']
    item_info.update({'_id':i['_id']},{'$set':{'area':area}})
```

```python
for item in item_info.find().limit(300):
    print(item['area'])
```

```python
pipeline = [
    {'$match':{'area':'朝阳'}},
    {'$group':{'_id':'$cates','counts':{'$sum':1}}},
    {'$sort':{'counts':-1}},
    {'$limit':3}
]
```

```python
index = []
post_times = []
for i in item_info.aggregate(pipeline):
    index.append(i['_id'])
    post_times.append(i['counts'])
```

```python
def data_gen(types):
    length = 0
    if length <= len(index):
        for cate, times in zip(index, post_times):
            data = {
                'name': cate,
                'data': [times],
                'type': types
            }
            yield data
            length += 1
```
