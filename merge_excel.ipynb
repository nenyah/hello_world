```python
import pandas as pd
import os
```

```python
def get_filenames(path):
	filenames = []
	for i in os.walk(path):
		for filename in i[-1]:
			full_filename = os.path.join(i[0],filename)
			filenames.append(full_filename)
	return filenames
```

```python
path = r'E:\Work\06-Work\00-Todo\Self Calc\Freight Self Calc Download'
print(get_filenames(path))
```

```python
def read_excel(filename):
	"""读入excel，返回dataFrame"""
	df = pd.read_excel(filename, index_col=None, headers = 0, na_values=['NA'])
	return df
```

```python
data = []
for filename in get_filenames(path):
    data.append(read_excel(filename))
print(data)
```

```python
def merge_excel(datas):
	"""合并数据"""
	return pd.concat(datas,ignore_index=True).drop_duplicates('ERP订单号')
```

```python
df = merge_excel(data)
df = df.reset_index()
df
```
