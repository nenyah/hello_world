```python
from pylab import *
from scipy.cluster.vq import *
import pandas as pd
import numpy as np
```

```python
path = r"./hanseless.csv"
```

```python
df = pd.read_csv(path, encoding="gbk")
```

```python
df.head()
```

```python
df.iloc[:,1:]
```

```python
data = vstack((df.iloc[:,1:].as_matrix()))
```

```python
data
```

```python
centroids,_=kmeans(data,4)
```

```python
result,_=vq(data,centroids) 
```

```python
result
```

```python
df['cat'] = pd.Series(result)
```
