```python
# 导入包
%pylab
import os
import pandas as pd
from pandas import Series,DataFrame
```

```python
def get_filenames(path):
	"""获取目录内所有文件名"""
	filenames = []
	for i in os.walk(path):
		for filename in i[-1]:
			full_filename = os.path.join(i[0],filename)
			filenames.append(full_filename)
	return filenames
```

```python
def read_excel(filename):
	"""读入excel，返回dataFrame"""
	df = pd.read_excel(filename, index_col=None, headers = 0, na_values=['NA'])
	return df
```

```python
def merge_excel(datas):
	"""合并数据"""
	return pd.concat(datas,ignore_index=True)
```

```python
path = r'E:\Work\06-Work\00-Todo\Order Rec'
files = get_filenames(path)
files
```

```python
datas = [*map(read_excel, files)]
len(datas)
```

```python
df = merge_excel(datas)
df.head()
```

```python
clean_df['sku_num'] = clean_df['sku'].str.split('<br/>')
clean_df.sku_num.str.len()
```

```python
single = clean_df[clean_df.sku_num.str.len() == 1.0]
single.info()
```

```python
sku_num = single.sku.str.extract('(?P<SKU>.*)\*(?P<order>\d)')
sku_num.head()
```

```python
single['SKU'] = sku_num.SKU
single.head()
```

```python
single['orders'] = [int(i) for i in sku_num.order]
single.head()
```

```python
# 定义函数，获得SPU
def get_spu(item):
    if item.startswith('B9'):
        spu,attr = item[:5],item[5:]
    elif item.startswith('EATGE'):
        spu,attr = item[:11],item[12:]
    elif item.startswith('WZ'):
        spu,attr = item[:9],item[9:]
    else:
        spu,attr = item[:7],item[7:]
        if attr.startswith('-'):
            attr = attr[1:]
    return spu,attr
```

```python
spu = pd.Series([i[0] for i in single.SKU.map(get_spu)])
attr = pd.Series([i[-1] for i in single.SKU.map(get_spu)])
single['SPU'],single['attr'] = spu,attr
single.iloc[:5,6:]
```

```python
# 按国家取TOP10
def top10(df,country='United States', startDate='2016-01-01', endDate='2016-08-31'):
    df = df[(startDate < df.paid_time) & (df.paid_time < endDate)]
    return df[df['country'] == country].groupby(['SPU']).sum().sort_values(by='orders', ascending=False)
```

```python
# 美国热销前10
us_top_10 = top10(single, 'United States', '2016-07-01', '2016-08-31')
us_top_10
```

```python
# 俄罗斯热销前10
ru_top_10 = top10(single, 'Russian Federation', '2016-08-01', '2016-08-31')
ru_top_10
```
